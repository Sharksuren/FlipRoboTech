Write a python program to access the search bar and search button on images.google.com and 
scrape 100 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’

```python
fruits=[]
cars=[]
Machine_learnings=[]
```

```python
key_words=['fruits','cars','Machine Learning']
url='https://images.google.com'

for i in key_words:
    driver=webdriver.Chrome(executable_path='C:\\Users\\suren\\FlipRobo\\Micro-Credit-Project\\chromedriver.exe')
    driver.get(url)
    driver.find_element(By.XPATH,'//input[@name="q"]').send_keys(i)
    driver.find_element(By.XPATH,'//button[@type="submit"]').click()
    
    img_ele=driver.find_elements(By.XPATH,'//img[@class="rg_i Q4LuWd"]')
    
    if i == 'fruits':
        for j in img_ele:
            while (len(fruits)<100):
                fruits.append(j.get_attribute("src"))
                
    elif i == 'cars':
        for j in img_ele:
            while (len(cars)<100):
                cars.append(j.get_attribute("src"))
                
    else:
        for j in img_ele:
            while (len(Machine_learnings)<100):
                Machine_learnings.append(j.get_attribute("src"))
    driver.close()
```

Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on 
www.flipkart.com and scrape following details for all the search results displayed on 1st page. 
Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, 
“Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Display 
Resolution”, “Processor”, “Processor Cores”, “Battery Capacity”, “Price”, “Product URL”. In 
case if any of the details is missing then replace it by “- “. Save your results in a dataframe 
and CSV.


```python
#creating web driver object
driver=webdriver.Chrome(executable_path='C:\\Users\\suren\\FlipRobo\\Micro-Credit-Project\\chromedriver.exe')

#defining url
url='https://www.flipkart.com/'

#initialising driver
driver.get(url)

#maximizing the window
driver.maximize_window()

#checking of the log in pop up is coming, if its comes closing the same
try:
    driver.find_element(By.XPATH,'//button[@class="_2KpZ6l _2doB4z"]').click()

except:
    print(' no pop up')

#entering the product to search
driver.find_element(By.XPATH,'//input[@title="Search for products, brands and more"]').send_keys('pixel 4A')

#clicking on search button
driver.find_element_by_xpath('//button[@class="L0Z3Pu"]').click()
```

```python

#creating empty list to store the scrapped data
Brand_name=[]
phone_name=[]
color=[]
RAM=[]
ROM=[]
Primary_camera=[]
Secondary_camera=[]
Display_size=[]
Display_resolution=[]
Processor=[]
Processor_core=[]
Battery_capacity=[]
Price=[]
product_url=[]

brand_ele=driver.find_elements(By.XPATH,'//div[@class="_4rR01T"]')
ram_ele=driver.find_elements(By.XPATH,'//ul[@class="_1xgFaf"]/li[1]')
dis_ele=driver.find_elements(By.XPATH,'//ul[@class="_1xgFaf"]/li[2]')
camera_ele=driver.find_elements(By.XPATH,'//ul[@class="_1xgFaf"]/li[3]')
battery_ele=driver.find_elements(By.XPATH,'//ul[@class="_1xgFaf"]/li[4]')
processor_ele=driver.find_elements(By.XPATH,'//ul[@class="_1xgFaf"]/li[5]')
price_ele=driver.find_elements(By.XPATH,'//div[@class="_30jeq3 _1_WHN1"]')
pro_url_ele=driver.find_elements(By.XPATH,'//a[@class="_1fQZEK"]')


for i in brand_ele:
    splitted_brand=i.text.split(' ')
    Brand_name.append(splitted_brand[0])
    
for i in brand_ele:
    splitted_phone=i.text.split('(')
    phone_name.append(splitted_phone[0])
    
for i in brand_ele:
    splitted_phone=i.text.split('(')
    splitted_color=splitted_phone[1].split(',')
    color.append(splitted_color[0])
    
for i in ram_ele:
    splitted_RAM=i.text.split('|')
    RAM.append(splitted_RAM[0])
    ROM.append(splitted_RAM[1])
    
for i in camera_ele:
    splitted_camera=i.text.split('|')
    Primary_camera.append(splitted_camera[0])
    if len(splitted_camera)==2:
        Secondary_camera.append(splitted_camera[1])
    else:
        Secondary_camera.append('--')
    
for i in dis_ele:
    splitted_display=i.text.split(')')
    Display_size.append(str(splitted_display[0]+')'))
    Display_resolution.append(splitted_display[1])
    
    
for i in processor_ele:
    if i.text is None:
        Processor.append('--')
        Processor_core.append('--')
    else:
        splitted_process=i.text.split(' ')
        Processor.append(str(splitted_process[0]+splitted_process[1]))
        if len(splitted_process)<4:
            Processor_core.append('--')
        else:
            Processor_core.append(str(splitted_process[2]+splitted_process[3]))
            
        #print(len(Processor_core))    
for i in battery_ele:
    Battery_capacity.append(i.text)
    
for i in price_ele:
    Price.append(i.text)
    
for i in pro_url_ele:
    product_url.append(i.get_attribute("href"))
```

```python
flip_phone=pd.DataFrame({})
flip_phone['Brand_name']=Brand_name
flip_phone['phone_name']=phone_name
flip_phone['color']=color
flip_phone['RAM']=RAM
flip_phone['ROM']=ROM
flip_phone['Primary_camera']=Primary_camera
flip_phone['Secondary_camera']=Secondary_camera
flip_phone['Display_size']=Display_size
flip_phone['Display_resolution']=Display_resolution
flip_phone['Processor']=Processor
flip_phone['Processor_core']=Processor_core
flip_phone['Battery_capacity']=Battery_capacity
flip_phone['Price']=Price
flip_phone['product_url']=product_url
```
