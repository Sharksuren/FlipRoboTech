```python
import pandas as pd
```

```python
df=pd.read_csv("Yatra_flight_price27.csv")
df.head()
```

```python
len(df)
```

```python
df.drop_duplicates()
```

```python
#Cleaning departure time as its scrapped with addition details
dept_time=[]
for time in df['departure_time']:
    split_time=time.split('\n')
    dept_time.append(split_time[0])  
```

```python
len(df)
```

```python
df['Departure_time']=dept_time
```

```python
#dropping old departure time

df.drop(columns=['departure_time','Unnamed: 0'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#cleanign the arrival time, removing +1 or 2 days as its added already in the duration
arrival=[]

for i in df['Arrival_time']:
    split=i.split('\n')
    arrival.append(split[0])
    
arrival
    
```

```python
df['Arrival_Time']=arrival
df.head()
```

```python
df.drop('Arrival_time',axis=1,inplace=True)
```

```python
#changing the departure date and time to date and time type

df['departure_date']=pd.to_datetime(df.departure_date)
df['Arrival_Time']=pd.DatetimeIndex(df.Arrival_Time).time
df['Departure_time']=pd.DatetimeIndex(df.Departure_time).time
```

```python
index_names = df[df['destination'] == 'Mumbai'].index
df.drop(index_names, inplace = True)
```
